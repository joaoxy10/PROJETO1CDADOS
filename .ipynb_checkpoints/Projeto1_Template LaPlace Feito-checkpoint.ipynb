## Montando SEU Classificador Naive-Bayes

Considerando apenas as mensagens da planilha Treinamento, ensine  seu classificador.

```python
import re 

def cleanup(text):
    """
        Função de limpeza muito simples que troca alguns sinais básicos por espaços
    """
    #import string
    #marcador='<br />'
    punctuation = '[´"!-.:?;$'']' # Note que os sinais [] são delimitadores de um conjunto.
    pattern = re.compile(punctuation)
    text_subbed = re.sub(pattern, '', text)
    return text_subbed
    
```

```python
def arruma_lista(serie_lista):
    serie2=[]
    for frase in serie_lista:
        frase_nova=cleanup(frase)
        frase_nova=frase_nova.replace("<br />"," ")
        frase_nova=frase_nova.replace("<br/>"," ")
        frase_nova=frase_nova.lower()
    #ler sobre str.replace(r'<br/s,regex=True)
        serie2.append(frase_nova)
    todas_palavras=[]
    for frase in serie2: 
        palavras=frase.split()
        for palavra in palavras:
            todas_palavras.append(palavra)
    serie_todas_palavras=pd.Series(todas_palavras)
    return serie_todas_palavras
```

```python
serie=pd.Series(train['Text']).tolist()
todas_palavras=arruma_lista(serie)
#serie_nova=cleanup(serie)

todas_palavras.value_counts().head(50)    
# olhar palvras que são indiferentes no classificador, como preposições
#stopwords
```

```python
train_0=train.loc[train['target']==0.0,'Text'].tolist()

todas_palavras_0=arruma_lista(train_0)
todas_palavras_0.value_counts().head(50) 
```

```python
train_1=train.loc[train['target']==1.0,'Text'].tolist()

todas_palavras_1=arruma_lista(train_1)
todas_palavras_1.value_counts().head(500) 
```

```python
train_2=train.loc[train['target']==2.0,'Text'].tolist()

todas_palavras_2=arruma_lista(train_2)
todas_palavras_2.value_counts().head(50) 
```

```python
freqr_tabela = todas_palavras.value_counts(True)  
freqr_tabela 
EP=todas_palavras.value_counts().sum()
```

```python
freqr_tabela_0 = todas_palavras_0.value_counts()
EP0=todas_palavras_0.value_counts().sum()
freqr_tabela_0 
```

```python
freqr_tabela_1 = todas_palavras_1.value_counts()
EP1=todas_palavras_1.value_counts().sum()
freqr_tabela_1 
```

```python
freqr_tabela_2 = todas_palavras_2.value_counts()  
EP2=todas_palavras_2.value_counts().sum()
freqr_tabela_2
```

```python
#def ClassNaiveBayes(frase)
p_0=EP0/EP
p_1=EP1/EP
p_2=EP2/EP
print(p_0,p_1,p_2)
```

## Verificando a performance do Classificador

Agora você deve testar o seu classificador com a base de Testes.

```python
def arruma_coluna(serie_lista):
    serie2=[]
    for frase in serie_lista:
        frase_nova=cleanup(frase)
        frase_nova=frase_nova.replace("<br />"," ")
        frase_nova=frase_nova.replace("<br/>"," ")
        frase_nova=frase_nova.lower()
    #ler sobre str.replace(r'<br/s,regex=True)
        serie2.append(frase_nova)
    return serie2
```

```python
teste=test.Text.tolist()
limpo=arruma_coluna(teste)
serie_limpo=pd.Series(limpo)
test['Text']=serie_limpo
test.head()
```

```python
#FUNÇÃO para cada palavra 

def prob_frase(frase):
    palavras = frase.split()
    
    prob_0, prob_1, prob_2 = p_0, p_1, p_2  
    
    for palavra in palavras:
        if palavra in freqr_tabela_0:
            prob_0 *= (freqr_tabela_0[palavra] + 1)/(EP0 + EP)
        else:
            prob_0 *= (0 + 1)/(EP0 + EP)

        if palavra in freqr_tabela_1:
            prob_1 *= (freqr_tabela_1[palavra] + 1)/(EP1 + EP)
        else:
            prob_1 *= (0 + 1)/(EP1 + EP)

        if palavra in freqr_tabela_2:
            prob_2 *= (freqr_tabela_2[palavra] + 1)/(EP2 + EP)
        else:
            prob_2 *= (0 + 1)/(EP2 + EP)

        if prob_0 <= 10**-300:
            prob_0 *= 10**50 

        if prob_1 <= 10**-300:
            prob_1 *= 10**50 

        if prob_2 <= 10**-300:
            prob_2 *= 10**50 

    
    total = prob_0 + prob_1 + prob_2
    return prob_0/total, prob_1/total, prob_2/total

#Listas com as probabilidades 
lista_p0 = []
lista_p1 = []
lista_p2 = []

for frase in test["Text"]:
    p0, p1, p2 = prob_frase(frase)   
    lista_p0.append(p0)
    lista_p1.append(p1)
    lista_p2.append(p2)

test["p_0"] = lista_p0
test["p_1"] = lista_p1
test["p_2"] = lista_p2


#Classe prevista 
preditos = []

for i in range(len(test)):
    valores = [test.loc[i, "p_0"], test.loc[i, "p_1"], test.loc[i, "p_2"]]
    classe = valores.index(max(valores))  # pega o índice do maior valor (0,1,2)
    preditos.append(classe)

test["predito"] = preditos

```

## Análise Qualitativa da Performance do Classificador

```python
true = [0, 0, 0]

false = [0, 0, 0]

for i in range(len(test.Text)):
    if test.target[i] == test.predito[i]:
        true[test.predito[i]] +=1
    elif test.target[i] != test.predito[i]:
        false[test.predito[i]] +=1


acuracia = sum(true)/len(test.Text)

total_0 = true[0] + false[0]
total_1 = true[1] + false[1]
total_2 = true[2] + false[2]

print(f'acuracia total = {100*acuracia:.3f}%')
print()
print(f'acuracia 0 = {100*true[0]/total_0:.3f}%')
print(f'acuracia 1 = {100*true[1]/total_1:.3f}%')
print(f'acuracia 2 = {100*true[2]/total_2:.3f}%')
print()
print(f'verdadeiros 0 = {true[0]}')
print(f'verdadeiros 1 = {true[1]}')
print(f'verdadeiros 2 = {true[2]}')
print()
print(f'falsos 0 = {false[0]}')
print(f'falsos 1 = {false[1]}')
print(f'falsos 2 = {false[2]}')
print()
```
